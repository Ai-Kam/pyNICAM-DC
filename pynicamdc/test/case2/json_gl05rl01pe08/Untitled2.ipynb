```python
import numpy as np
import matplotlib.pyplot as plt

# Define the file path (Update this with the correct local path)
file_path = "vgrid40_stretch_45km.dat"  # Ensure the file is in the same directory or provide full path

# Read the file as big-endian float64 data
data = np.fromfile(file_path, dtype=">f8")  # Read as big-endian float64

# Plot the data
plt.figure(figsize=(8, 5))
plt.plot(data, marker="o", linestyle="-", color="b", markersize=4)

# Labels and title
plt.title("Vgrid40 Stretch 45km Data")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True)

# Show the plot
plt.show()
```
