```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Define the JSON file path (Update this if needed)
json_file_path = "vgrid40_stretch_45km.json"

# Load the JSON file
with open(json_file_path, "r") as json_file:
    data = json.load(json_file)

# Convert lists back to NumPy arrays
set1 = np.array(data["set1"])
set2 = np.array(data["set2"])

# Create a figure
plt.figure(figsize=(8, 5))

# Plot both sets
plt.plot(set1, marker="o", linestyle="-", color="b", markersize=4, label="Set 1")
plt.plot(set2, marker="s", linestyle="--", color="r", markersize=4, label="Set 2")

# Labels and title
plt.title("Vgrid40 Stretch 45km Data")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```
